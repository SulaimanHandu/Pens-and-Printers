# Pens and Printers Sales Report 
### **Company Overview**

Founded in 1984, Pens and Printers is a trusted provider of high-quality office products, ranging from pens and notebooks to desks, chairs, and monitors. As consumer purchasing behaviors evolve, so must our sales tactics. Launching a new product line is a significant investment, and it’s crucial to employ the most effective strategies to ensure success.

### **Data Preparation**

We received data from the sales team to analyse our sales performance, which was extracted from their tracking system. This data will serve as the foundation for our analysis, guided by the information provided in the metadata. Our first step was to validate and clean the data using Power BI.

### **Data Analysis**

A pilot was conducted with 15,000 customers to evaluate the effectiveness of various sales methods. Among the methods tested, email alone reached the highest number of customers, while the combination of email and calls engaged the fewest. The chart below provides a detailed breakdown.

### **Revenue Insights**

The total revenue generated from this pilot was approximately 1.31 million. As anticipated, the email-only method produced the highest revenue, correlating with the larger customer base engaged through this channel. However, despite reaching fewer customers, the email and call combination outperformed the calling-only method, generating 80% more revenue. This highlights the effectiveness of the combined approach.

### **Performance Over Time**

The pilot spanned six weeks, during which the different sales methods exhibited varied performance. The calling method maintained a relatively constant sales rate across all weeks, while the email method saw a strong initial response in the first week, followed by a steady decline. On the other hand, the email and call combination started slow but showed significant improvement over time, indicating its potential as a long-term strategy.

![Screenshot 2024-08-23 011001](Screenshot%202024-08-23%20011001.png)

 
### **Customer Engagement and Revenue Per Customer**

To account for the varying customer numbers across methods, we analyzed the average revenue per customer. The results revealed that the email and call combination was significantly more effective, generating 77% more revenue per customer than email alone and 246% more than calls alone.
 

![Screenshot 2024-08-23 004707](Screenshot%202024-08-23%20004707.png)


### **Customer Tenure and Purchase Behavior**

Customers were categorized based on their tenure with us. The analysis suggests that newer customers are more likely to purchase products than older customers, as indicated by a right-skewed distribution.

### **Site Visits and Product Sales**

When comparing the total number of site visits and products sold across customer categories, we observed consistent behavior. Customers who received both calls and emails visited the site more frequently, reinforcing the effectiveness of the combined sales approach.

To estimate the total revenue projection for the 15,000 customers based on the average revenue of 159 per customer from the Email + Call method, you can use the following calculation:

2. **Total Number of Customers**: 15,000

So, the projected total revenue for 15,000 customers, based on the average revenue of 159 per customer from the Email + Call method over the same period of time, would be 2,385,000. This is about 77% increase in revenue compared to what we have in the current data. 
 
### **Conclusion**

In conclusion, the analysis reveals that while email alone reached the most customers and generated the highest overall revenue, the combination of email and calls proved to be the most effective method in terms of revenue per customer. Despite initially slow performance, the combined approach showed significant improvement over time, suggesting its potential as a powerful long-term sales strategy. Additionally, the analysis highlighted that newer customers are more inclined to purchase products, emphasizing the need to tailor our marketing efforts based on customer tenure. Overall, the combined method of email and calls emerged as the most promising tactic for driving sustained revenue growth.

### **Recommendation**

Based on the analysis and the nature of each sales method, I recommend prioritizing the Email + Call approach for future sales campaigns. While the Email-only method requires minimal effort, it lacks the personalized touch that can drive higher engagement. The Call-only method is resource-intensive, with sales team members spending significant time per customer, yet it doesn't yield the highest revenue.
The Email + Call method strikes a balance by leveraging the efficiency of email communication and the effectiveness of personalized follow-up calls. The 10-minute calls in this approach are more time-efficient than the 30-minute calls used in the Call-only method, yet they still provide valuable customer interaction. This method not only generated significantly higher revenue per customer but also showed increasing effectiveness over time, making it a strong long-term strategy. By adopting the Email and Call approach, Pens and Printers can maximize revenue while optimizing the use of sales team resources.
